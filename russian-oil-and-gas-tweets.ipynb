```python
import twint

c = twint.Config()

c.Search = ['(Russia) AND (Oil OR Gas)']# topic
c.min_likes=500
c.Limit = 5000      # number of Tweets to scrape
c.Store_csv = True       # store tweets in a csv file
c.Output = "Tweets_Russian_Oil_Gas.csv"     # path to csv file

twint.run.Search(c)
```
